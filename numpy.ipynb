```python
np.random.seed(25)

df1_obj = DataFrame(np.random.rand(36).reshape((6,6)),
                    index=['row 1','row 2','row 3','row 4','row 5','row 6'], 
                   columns=['column 1','column 2','column 3','column 4','column 5','column 6'])

df1_obj
```

```python
df1_obj.loc['row 2':'row 4', 'column 1'] = missing
df1_obj.loc['row 1':'row 3', 'column 6'] = missing


df1_obj
```

```python
filled_df = df1_obj.fillna(0)
filled_df
```

```python
# Replacing nan with values 0.25 and 1.25
Filled_df = df1_obj.fillna( value = {"column 1": 0.25, "column 6": 1.25})
Filled_df
```

```python
#Missing replaced with the last filled value backward
fill_df = df1_obj.fillna(method = "bfill")
fill_df
```

```python
# Filtering out missing values using columns
DF_no_NaN = df1_obj.dropna(axis=1)
DF_no_NaN
```
