Creating statistical charts

```python
import numpy as np
import pandas as pd
import chart_studio.plotly as py 
import chart_studio.tools as tls
import plotly.graph_objs as go 
import cufflinks as cf

import sklearn
from sklearn.preprocessing import StandardScaler

```

```python
address = 'C:/Users/HP/Documents/Programming/PROJECTS/Exercise Files/Data/mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars.mpg

mpg.iplot(kind='histogram',filename='simple-histogram')
```

Multiple histogram

```python
cars_subset=cars[['mpg','disp','hp']]

cars_data_std = StandardScaler().fit_transform(cars_subset)

cars_select = pd.DataFrame(cars_data_std)
cars_select.columns = ['mpg','disp','hp']

cars_select.iplot(kind='histogram',filename='Multiple-histogram-chart')
```

Subplot Histograms

```python
cars_select.iplot(kind='histogram',subplots=True,filename='Subplot-histogram-chart')
```

Subplot histogram horizontal

```python
cars_select.iplot(kind='histogram',filename='subplots-hist',subplots=True,shape=(3,1))
```

Subplot histogram vertical 

```python
cars_select.iplot(kind='histogram',filename='subplots-hist',subplots=True,shape=(1,3))
```

Creating a Boxplot

```python
cars_select.iplot(kind='box',filename='Box-plot-chart')
```

Creating Scatter plot

```python
fig = {'data':[{
    'x': cars_select.mpg,'y':cars_select.disp,'mode':'markers','name':'mpg'},
    {'x': cars_select.hp,'y':cars_select.disp,'mode':'markers','name':'hp'}],
    'layout':{'xaxis':{'title':''},'yaxis':{'title':'Standardized Displacement'}}}

py.iplot(fig,filname='grouped-scatter-plot')
```
