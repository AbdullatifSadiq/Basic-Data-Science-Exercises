Defining axes,ticks and grids

Creating labels using the Object orientated method.

```python
address = "C:/Users/HP/Documents/Programming/PROJECTS/Exercise Files/Data/mtcars.csv"

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')

ax.set_title('Miles per Gallon of cars from mtcars dataset')

ax.set_xlabel(' Car names')
ax.set_ylabel(' mile/gal')
```

Creating a legend using the OO method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')

ax.set_title('Miles per Gallon of cars from mtcars dataset')

ax.set_xlabel(' Car names')
ax.set_ylabel(' mile/gal')

ax.legend(loc='best')
```

Annotating plots

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,4
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')

ax.set_title('Miles per Gallon of cars from mtcars dataset')

ax.set_xlabel(' Car names')
ax.set_ylabel(' mile/gal')

ax.legend(loc='best')


ax.set_ylim([0,45])

ax.annotate('Toyota Corolla',xy=(19,33.9),xytext=(21,35),
            arrowprops=dict(facecolor='black',shrink=0.05))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from pylab import rcParams
from pandas.plotting import scatter_matrix



import seaborn as sb
sb.set_style('whitegrid')
```

Creating a histogrm using the mrcars dataset

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4


address = "C:/Users/HP/Documents/Programming/PROJECTS/Exercise Files/Data/mtcars.csv"

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']



cars.index = cars.car_names

mpg = cars['mpg']

mpg.plot(kind='hist')
plt.xlabel('x axis')
plt.ylabel('y axis')
```
